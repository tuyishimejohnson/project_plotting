```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("fuel_econ.csv")

# Creating a histogram for carbon dioxid column
plt.hist(df["co2"], bins=15, color="green",edgecolor="white")

plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.title("Distribution of CO2 Emissions")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Reading data from the file.
df = pd.read_csv('fuel_econ.csv')

# selecting numeric columns
numerics = df.select_dtypes(include=['float64', 'int64'])

# width and height of the heatmap
plt.figure(figsize=(20,10))
sns.heatmap(numerics.corr())
sns.heatmap(numerics.corr(),annot=True)
```
